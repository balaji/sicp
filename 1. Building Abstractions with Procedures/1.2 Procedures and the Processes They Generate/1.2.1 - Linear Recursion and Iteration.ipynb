### Exercise 1.9: 
Each of the following two procedures defines a method for adding two positive integers in terms of the procedures inc, which increments its argument by 1, and dec, which decrements its argument by 1.
```scheme
(define (+ a b)
  (if (= a 0) b (inc (+ (dec a) b))))

(define (+ a b)
  (if (= a 0) b (+ (dec a) (inc b))))
```
Using the substitution model, illustrate the process generated by each procedure in evaluating (+ 4 5). Are these processes iterative or recursive?


#### Answer:
The first definition of the addition procedure is *linear-recursive*. i.e., the stack grows linearly with `n`.
```scheme
(+ 5 4)
(inc (+ (dec 5) 4))
(inc (inc (+ (dec 4) 4)))
(inc (inc (inc (+ (dec 3) 4))))
(inc (inc (inc (inc (+ (dec 2) 4)))))
(inc (inc (inc (inc (inc (+ (dec 1) 4))))))
(inc (inc (inc (inc (inc (+ 0 4))))))
(inc (inc (inc (inc (inc 4)))))
(inc (inc (inc (inc 5))))
(inc (inc (inc 6)))
(inc (inc 7))
(inc 8)
9
```

The second definition of the addition process is *linear-iterative*. i.e. the number of steps to execute grows linearly with `n`, while the stack remains a constant.
```scheme
(+ 5 4)
(+ (dec 5) (inc 4))
(+ (dec 4) (inc 5))
(+ (dec 3) (inc 6))
(+ (dec 2) (inc 7))
(+ (dec 1) (inc 8))
(+ 0 9)
9
```

### Exercise 1.10: 
The following procedure computes a mathematical function called Ackermannâ€™s function.
```scheme
(define (A x y)
  (cond ((= y 0) 0)
        ((= x 0) (* 2 y))
        ((= y 1) 2)
        (else (A (- x 1) (A x (- y 1))))))
```

What are the values of the following expressions?
```scheme
(A 1 10)
(A 2 4)
(A 3 3)
```
Consider the following procedures, where A is the procedure defined above:
```scheme
(define (f n) (A 0 n))
(define (g n) (A 1 n))
(define (h n) (A 2 n))
(define (k n) (* 5 n n))
```
Give concise mathematical definitions for the functions computed by the procedures `f`, `g`, and `h` for positive integer values of *n*. For example, `(k n)` computes `5n^2`.

#### Answer:
Given the function `A`

````racket
(define (A x y)
  (cond ((= y 0) 0)
        ((= x 0) (* 2 y))
        ((= y 1) 2)
        (else (A (- x 1) (A x (- y 1))))))
````

The values of the following expressions are:

````racket
(A 1 10)
````

````racket
(A 2 4)
````

````racket
(A 3 3)
````

The concise mathematical definitions for the following functions are:

```scheme
(define (f n) (A 0 n))
(f 1) ; = 2
(f 2) ; = 4
(f 3) ; = 6
```
=> $f(n) = 2n$

```scheme
(define (g n) (A 1 n))
(g 1) ;= 2
(g 2) ;= 4
(g 3) ;= 8
```
=> $g(n) = 2^{n}$

```scheme
(define (h n) (A 2 n))
(h 0) ;= 0
(h 1) ;= 2 = (2^0)
(h 2) ;= 4 = (2^2)
(h 3) ;= 16 = (2^4)
(h 4) ;= 65536 = (2^16)
(h 5) ;= ... = (2^65536)
```
=> $h(n) = 2^{h(n-1)}$
